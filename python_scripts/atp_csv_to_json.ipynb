```python
with open('../assets/content/species_info_vi.json', 'r') as f:
    data = json.load(f)
df = pd.read_csv('species_atp_vi.csv', encoding='utf-8')

for species in data.keys():
    data[species]['primary_name'] =  df.loc[df['Họ/Loài'] == species]['Tên thường gọi'].values[0]
    data[species]['primary_name'] =  df.loc[df['Họ/Loài'] == species]['Tên thường gọi'].values[0]
    data[species]['primary_name'] =  df.loc[df['Họ/Loài'] == species]['Tên thường gọi'].values[0]
```

```python
def csv_to_json_vi(csv_path, json_path):
    df = pd.read_csv(csv_path, encoding='utf-8')

    data = {"species": {}}
    for _, row in df.iterrows():
        data['species'][row['Họ/Loài']] = \
                {
                    'scientific_name': row['Họ/Loài'],
                    'primary_name': row['Tên thường gọi'],
                    'iucn': row['Hiện trạng bảo tồn'],
                    'sdvn': row['Luật pháp bảo vệ'],
                    'law': "",
                    'secondary_name': '',
                    'distribution': '',
                    'habitat': '',
                    'reference_images': [],
                    'identification_images': [],
                    'fun_facts': ''
                }
        # data['species'].append(item)

    with open(json_path, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

csv_to_json_vi('species_atp_vi.csv', '../assets/content/species_info_vi.json')
```
